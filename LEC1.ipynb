```python
import pandas as pd
import numpy as np
```

```python
#1.Missing Values
data={'Name':['Alice','Bob',np.nan,'David'],'Age':[25,np.nan,30,22],'Salary':[50000,54000,np.nan,48000]}
```

```python
df=pd.DataFrame(data)

print("Original Data:\n",df)
```

```python
df_cleaned=df.dropna()

print("Data after removing missing values:\n",df_cleaned)
```

```python
#2.Handling Duplicates
data2={'Name':['Alice','Bob','Alice','David'],'Age':[25,30,25,22],'Salary':[50000,54000,50000,48000]}
```

```python
df2=pd.DataFrame(data2)
print(df2)
```

```python
df_duplicates=df2.drop_duplicates()
print(df_duplicates)
```
