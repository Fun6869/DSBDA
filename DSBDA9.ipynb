```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Filling null values
titanic['age']=titanic['age'].fillna(titanic['age'].mean())
titanic['embarked']=titanic['embarked'].fillna(titanic['embarked'].mode()[0])
titanic['deck']=titanic['deck'].fillna(titanic['deck'].mode()[0])
titanic['embark_town']=titanic['embark_town'].fillna(titanic['embark_town'].mode()[0])
```

```python
titanic.isnull().sum()
```

```python
# Drop rows with missing 'age' or 'sex' or 'survived' to avoid errors
titanic = titanic.dropna(subset=['age', 'sex', 'survived'])
```

```python
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
plt.title('Age Distribution by Gender and Survival')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.countplot(x='sex',hue='survived',data=titanic)
plt.title("survival based on gender")
plt.show()
```

```python
sns.histplot(titanic['age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Age vs Fare Colored by Survival')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Survival Count Based on Passenger Class')
plt.show()
```
